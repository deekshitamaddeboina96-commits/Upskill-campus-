```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('/content/train_aWnotuB.csv')
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.plot(x='DateTime', y='Vehicles', kind='line')

plt.title("Smart City Traffic Pattern Analysis")
plt.xlabel("Date and Time")
plt.ylabel("Number of Vehicles")

plt.show()
```

```python
top5 = df.groupby('Junction')['Vehicles'].mean().sort_values(ascending=False).head(5)

top5.plot(kind='bar')

plt.title("Top 5 High Traffic Junctions")
plt.xlabel("Junction")
plt.ylabel("Average Vehicles")

plt.xticks(rotation=45)

plt.show()
```

```python
traffic = df['Vehicles']

plt.plot(traffic)

plt.title("Smart City Traffic Trend")
plt.xlabel("Time")
plt.ylabel("Number of Vehicles")

plt.show()
```
